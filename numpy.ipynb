time taken by a list 
Process_time() function finds the time to complete the range 


```python
from time import process_time
python_list=[i for i in range(10000)]
start_time=process_time() #process_time function finds the time taken to complete the range 
python_list= [i+5 for i in python_list] #adding the value 5
end_time= process_time()
print(end_time- start_time )
```

```python

from time import process_time
import numpy as np

# Create a NumPy array
np_array = np.array([i for i in range(10000)])

# Start timer
start_time = process_time()

# Add 5 to every element
np_array += 5

# End timer
end_time = process_time()

# Print CPU time taken
print("Time taken:", end_time - start_time, "seconds")
```
