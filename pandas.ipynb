```python
#1.CREATE A DATAFRAME
import pandas as pd

data = {
    "name": ["alice", "bob"],
    "age": [20, 10],
    "salary": [50000, 80000]
}

df = pd.DataFrame(data)
print(df)
```

```python
#read data from csv file through different methods
#CSV :
df=pd.read_csv("data.csv")

#EXCEL: 
df=pd.read_excel("data.xlsx")

#JSON:
df=pd.read_json("data.json")
```

```python
import pandas as pd

data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [25, 35, 30, 40],
    "gender": ["female", "male", "male", "male"]
}

df = pd.DataFrame(data)
print(df)
#filtering data

#single condition
df[df["age"] > 30]

#multiple conditions
df[(df["age"] > 30) & (df["gender"] == "male")]
```

```python
#7 add a column 
import pandas as pd

data = {
    "salary": [100, 200, 300]
}

df = pd.DataFrame(data)

df["bonus"] = df["salary"] * 0.10

print(df)
```
